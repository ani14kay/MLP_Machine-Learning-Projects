 # #ML-Mini-6 : Digit Recognition using Random Forest

### 1 - Importing Libraries

```python
import pandas as pd
import numpy as np 
```

### 2 - Uploading the Dataset

```python
dts = pd.read_csv("Digits.csv")
dts
```

### 3 - Summarizing the Dataset

```python
dts.shape
```

```python
dts.head()
```

### 5  - Segregating Dataset into X [ Input / Independant Variable ] & Y [ Output / Dependant Variable ]

```python
X = dts.iloc[:,1:]
X
```

```python
X.shape
```

```python
Y = dts.iloc[:,0]
Y
```

```python
Y.shape
```

### 6 - Splitting Dataset to Test & Train

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

### 9 - Training our Model

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

### 12 - Prediction Results for a sample value

```python
import matplotlib.pyplot as plt
index=96
print("Predicted : "+str(model.predict(X_test)[index]))
plt.axis('off')
plt.imshow(X_test.iloc[index].values.reshape((28,28)),cmap='gray')
```

### 13 - Evaluating Accuracy of the Model

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model is {0}%".format(accuracy_score(Y_test,Y_pred)*100))
```
